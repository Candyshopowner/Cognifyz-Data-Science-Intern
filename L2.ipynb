```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset .csv")
df
```

```python
query1 = df[(df['Has Online delivery']=='Yes') & (df["Has Table booking"]=='Yes')]
query1
```

```python
percent = (len(query1)/len(df))*100
```

```python
print("Percentage of restaurants which offer Table Booking and Online Delivery is :",percent,"%")
```

```python
rating_witb = df['Aggregate rating'][df['Has Table booking']=='Yes']
rating_wotb = df['Aggregate rating'][df['Has Table booking']=='No']
```

```python
avg_witb = rating_witb.mean()
avg_wotb = rating_wotb.mean()
```

```python
print("Average rating of restaurant with Table Booking is :",avg_witb)
print("Average rating of restaurant without Table Booking is :",avg_wotb)
```

```python
online_del = df[df['Has Online delivery']=='Yes']
online_del
```

```python
sns.displot(x='Price range',data=online_del,bins=10,kde=True)
plt.title("Availability of online delivery among restaurants with different price ranges.")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.displot(data=df,x='Price range',bins=10,kde=True)
plt.title("Distribution of Price Range")
plt.show()
```

```python
com_price_range = df["Price range"].value_counts().reset_index()
com_price_range
```

```python
print("Most Common Price Range is :",com_price_range.value_counts().idxmax())
```

```python
avg_rating = {}
for x in df["Price range"].unique():

   avg_rating[x] = df['Aggregate rating'][df['Price range']==x].mean()
```

```python
for item in avg_rating:
    print("Average rating of Price Range {} is :".format(item),avg_rating[item])
```

```python
print("The color representing the highest rating for all price ranges is Dark Green")
```
